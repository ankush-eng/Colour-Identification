### Importing Required Libraries 

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
img1_path = 'colorpic.jpg' 
img2_path = 'lena.jpg' 
csv_path = 'colors.csv'
```

```python
img2 = cv2.imread(img2_path)
img1 = cv2.imread(img1_path)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(img1)
```

```python
index = ['colour', 'colour_name', 'hex', 'R', 'G', 'B']
df = pd.read_csv(csv_path, names=index, header=None)
```

```python
clicked = False
r = g = b = xpos = ypos = 0
```

```python
def get_color_name(R,G,B):
    minimum = 1000
    for i in range (len(df)):
        d = abs(R - int(df.loc[i,'R'])) + abs(G - int(df.loc[i,'G'])) + abs(B - int(df.loc[i,'B'])) 
        if d <= minimum:
            minimum = d
            cname = df.loc[i,'colour_name']   
    return cname        
```

```python
def draw_function(event, x, y, flags, params):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        global b,g,r, xpos, ypos, clicked
        clicked = True
        xpos = x
        ypos = y
        b,g,r = img1[y,x]
        b = int(b)
        g = int(g)
        r = int(r)
```

```python
cv2.namedWindow('Task 2')
cv2.setMouseCallback('Task 2', draw_function)
```

```python
while True:
    cv2.imshow('Task 2',img1)
    if clicked:
        cv2.rectangle(img1, (20,20), (600,60), (b,g,r), -1)
        text = get_color_name(r,g,b) + 'R=' + str(r) + 'G=' + str(g) + 'B=' + str(b)
        cv2.putText(img1, text, (50,50), 2, 0.8, (255,255,255), 2, cv2.LINE_AA)
        
    if cv2.waitKey(20) & 0xFF == 27:
             break
            
cv2.destroyAllWindows()        
```
